# Adding Exception handling

```python
try:
    input_two = int(input("Enter value: "))
    print(f"input_two: {input_two}")
except ValueError:
    print("Invalid input")
```

```python
try:
    input_three = int(input("Enter value: "))
except ValueError:
    print("Invalid input")
else:
    print(f"input_three: {input_three}")
```

```python
condition = True
while condition:
    try:
        input_three = int(input("Enter value: "))
    except ValueError:
        print("Invalid input")
        condition = True
    else:
        print(f"input_three: {input_three}")
        condition = False
```

```python
while True:
    try:
        input_three = int(input("Enter value: "))
    except ValueError:
        print("Invalid input")
    else:
        print(f"input_three: {input_three}")
        break
```

```python
while True:
    try:
        input_three = int(input("Enter value: "))
        break
    except ValueError:
        print("Invalid input")
print(f"input_three: {input_three}")
```

```python
def get_num():
    while True:
        try:
            input_four = int(input("Enter value: "))
        except ValueError:
            print("Invalid input")
        else:
            break
    return input_four
```

```python
def get_num():
    while True:
        try:
            return int(input("Enter value: "))
        except ValueError:
            print("Invalid input")
```
