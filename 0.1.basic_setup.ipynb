# Removing un-wanted space

```python
name = input("What is your name ?")
#Remove additional space
name = name.strip()
#name = name.capitalize()
name = name.title()
print("my name ",name, sep=": ", end="\nEND")
```

# Chaining the functions

```python
name = input("What is your name ?").strip().title()
#Remove additional space and capitalize
#name = name.strip().title()
print("My name ", name, sep=": ", end="\nEND")
# Chaining the functions
```

# Splitting data

```python
name = input("What is your name ?").strip().title()
first, last = name.split(" ")
print("My name ", first, last, sep=": ", end="\nEND")
```
