# random

## choice

```python
import random
from random import choice
conisn = choice(["Good", "Bad"])
print(conisn)
```

```python
import random
from random import randint
conisn = randint(0, 10)
print(conisn)
```

## shuffle

```python
from random import shuffle
cards = ["King", "Ace", "Queen", "Jack", "King", "Ace", "Queen", "Jack"]
shuffle(cards)
print(cards)
```
