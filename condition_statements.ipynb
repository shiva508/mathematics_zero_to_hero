# Conditions
## if
## elif
## else

```python
from re import match

input_one = int(input("Enter a number: "))
input_two = int(input("Enter another number: "))
if input_one > input_two:
    print(f"{input_one} is greater than {input_two}")
elif input_one < input_two:
    print(f"{input_one} is less than {input_two}")
else:
    print(f"{input_one} is equal to {input_two}")
```

# OR

```python
input_one = int(input("Enter a number: "))
input_two = int(input("Enter another number: "))
if input_one > input_two or input_one < input_two:
    print(f"{input_one} is not equal to {input_two}")
else:
    print(f"{input_one} is equal to {input_two}")
```

# != or ==

```python
input_one = int(input("Enter a number: "))
input_two = int(input("Enter another number: "))
if input_one == input_two:
    print(f"{input_one} is equal to {input_two}")
else:
    print(f"{input_one} is not equal to {input_two}")
```

# and

```python
score = int(input("Enter a score: "))
if score >= 90 and score <= 100:
    print("Grade : A")
elif score >= 80 and score <= 90:
    print("Grade : B")
elif score >= 70 and score <= 80:
    print("Grade : C")
elif score >= 60 and score <= 70:
    print("Grade : D")
else:
    print("Grade : F")
```

```python
score = int(input("Enter a score: "))
if 90 <= score <= 100:
    print("Grade : A")
elif 80 <= score <= 90:
    print("Grade : B")
elif 70 <= score <= 80:
    print("Grade : C")
elif 60 <= score <= 70:
    print("Grade : D")
else:
    print("Grade : F")
```

# +, -, *, %

```python
def is_even_can_be_improved(num):
    if num % 2 == 0:
        return True
    else:
        return False
def is_even_good(num):
    return True if num % 2 == 0 else False

def is_even_better(num):
    return num % 2 == 0

x = int(input("Enter a number: "))
if is_even_better(x):
    print(f"{x} is even")
else:
    print(f"{x} is odd")
```

# match

```python
name = input("Enter a name: ")
match name:
    case "Batman":
        print("Welcome to Protagonist")
    case "Joker":
        print("Welcome to Antagonist")
    case "Superman":
        print("Welcome to Protagonist")
    case "Flash":
        print("Welcome to Protagonist")
    case "Cyberg":
        print("Welcome to Protagonist")
    case "Dark Shed":
        print("Welcome to Antagonist")
    case _:
        print("Please enter a valid name")
```

```python
name = input("Enter a name: ")
match name:
    case "Batman" | "Superman" | "Flash" | "Cyberg" | "Wonder Woman" :
        print("Welcome to Protagonist")
    case "Joker" | "Dark Shed":
        print("Welcome to Antagonist")
    case _:
        print("Please enter a valid name")
```
