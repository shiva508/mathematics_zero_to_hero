# Writing data to file

```python
condition = True
names = []
while condition:
    name = input("Please enter your name: ")
    if name == "exit":
        condition = False
    else:
        names.append(name)
        print(name)

print(names)
with open("names.txt", "a") as file:
    for name in names:
        file.write(name + "\n")
```

# Read file data

```python
with open("names.txt", "r") as file:
    lines = file.readlines()

for line in lines:
    print(line, end="")
```

```python
full_names = []
with open("names.csv", "r") as file:
    for line in file:
        last_name, first_name = line.rstrip().split(",")
        ful_name = {
            "first_name": first_name,
            "last_name": last_name
        }
        full_names.append(ful_name)
        #full_names.append(f"{last_name} {first_name}")
def get_first_name(full_name):
    return full_name["first_name"]
for full_name in sorted(full_names, key=get_first_name, reverse=True) :
    print(full_name)
```
